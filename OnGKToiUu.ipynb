## Gradient Descent

$$\displaystyle \begin{align*}
x^{(k)} &=x^{(k-1)} - t . \nabla f_{x^{(k-1)}}\\
&=x^{(k-1)} - \frac{1}{\delta + 1}(2x_1^{(k-1)},2.\delta.x_2^{(k-1)}) \\
&=(x_1^{(k-1)},x_2^{(k-1)}) - (\frac{2}{\delta+1}x_1^{(k-1)},\frac{2\delta}{\delta+1}x_2^{(k-1)}) \\
&= (\frac{\delta-1}{\delta+1}x_1^{(k-1)},\frac{1-\delta}{\delta+1}x_2^{(k-1)}) \\ 
\end{align*}$$



$$\displaystyle \begin{align*}
x_1^{(k)} = \frac{\delta-1}{\delta+1}x_1^{(k-1)} = \frac{\delta-1}{\delta+1}. \frac{\delta-1}{\delta+1} x_1^{(k-2)} = \dots = (\frac{\delta-1}{\delta+1})^k . x_1^{(0)} = (\frac{\delta-1}{\delta+1})^k . \delta 

\end{align*}$$

$$\displaystyle \begin{align*}
x_2^{(k)} = \frac{1-\delta}{\delta+1}x_2^{(k-1)} = \dots = (\frac{1-\delta}{\delta+1})^k . x_2^{(0)} = (\frac{\delta-1}{\delta+1})^k 
\end{align*}$$

Vy $$\displaystyle x^{(k)} = \left((\frac{\delta-1}{\delta+1})^k.\delta \quad,\quad (\frac{\delta-1}{\delta+1})^k \right)$$

$f(x) = x^2-2x+5$ 

$f'(x) = 2x-2$

````python
import numpy as np
import math
import matplotlib.pyplot as plt
````

````python
import numpy as np

def grad(x):
    return 2 * x - 2

def cost(x):
    return x**2 - 2*x + 5

def myGD1(eta, x0):
    x = [x0]
    for it in range(100):
        x_new = x[-1] - eta * grad(x[-1])
        if abs(grad(x_new)) < 1e-3:
            break
        x.append(x_new)
    return x, it

# p dng thut ton GD  gii bi ton
(x1, it1) = myGD1(0.1, -5) #eta l learning rate t
(x2, it2) = myGD1(0.1, 5)

# In kt qu
print('Solution x1 = %f, cost = %f, obtained after %d iterations' % (x1[-1], cost(x1[-1]), it1))
print('Solution x2 = %f, cost = %f, obtained after %d iterations' % (x2[-1], cost(x2[-1]), it2))
````

````python
x = np.linspace(-6, 6, 100)
y = cost(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x)')
plt.scatter(x1, [cost(xi) for xi in x1], c='r', marker='o', label='GD path (x1)')
plt.scatter(x2, [cost(xi) for xi in x2], c='g', marker='s', label='GD path (x2)')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent Optimization')
plt.grid(True)
plt.show()
````

Gradient Descent (GD) cng c th c p dng cho cc hm nhiu bin, tc l hm m c nhiu hn mt bin u vo. Khi ti u ha hm nhiu bin bng GD, chng ta s tnh gradient ca hm  i vi tng bin ring l v cp nht cc bin theo hng m ca gradient.

Di y l mt v d v cch p dng GD cho mt hm hai bin:

Hm mc tiu: $f(x, y) = x^2 + y^2$

Gradient ca hm mc tiu: $\nabla f(x, y) = \begin{bmatrix}2x \\ 2y\end{bmatrix}$

Thut ton GD cho hm hai bin:

1. Chn gi tr khi to: $(x_0, y_0)$.
2. Lp qua cc bc sau cho n khi tiu ch dng (v d: gradient nh hn mt ngng nht nh):
   - Tnh gradient ti im hin ti: $\nabla f(x_i, y_i) = \begin{bmatrix}2x_i \\ 2y_i\end{bmatrix}$.
   - Cp nht $(x, y)$ theo hng m ca gradient:
     $$x_{i+1} = x_i - \eta \cdot \frac{\partial f}{\partial x}(x_i, y_i) = x_i - 2\eta x_i$$
     $$y_{i+1} = y_i - \eta \cdot \frac{\partial f}{\partial y}(x_i, y_i) = y_i - 2\eta y_i$$
   - Tng s lng lp ln: $i = i + 1$.

Di y l mt v d m Python thc hin GD cho hm hai bin $f(x, y) = x^2 + y^2$:

M ny s ti u ha hm $f(x, y) = x^2 + y^2$ bng GD v in ra gi tr ti u ca $(x, y)$ cng vi gi tr ca hm mt mt v s ln lp.

````python
import numpy as np

def gradient(x, y):
    grad_x = 2 * x
    grad_y = 2 * y
    return grad_x, grad_y

def cost(x, y):
    return x**2 + y**2

def gradient_descent(eta, x0, y0, max_iter=100, epsilon=1e-5):
    x = [x0]
    y = [y0]
    for i in range(max_iter):
        grad_x, grad_y = gradient(x[-1], y[-1])
        x_new = x[-1] - eta * grad_x
        y_new = y[-1] - eta * grad_y
        x.append(x_new)
        y.append(y_new)
        if np.sqrt(grad_x**2 + grad_y**2) < epsilon:
            break
    return x, y, i

# p dng GD  ti u ha hm f(x, y)
x0, y0 = 3, 4  # Gi tr khi to
eta = 0.1  # Learning rate
x, y, it = gradient_descent(eta, x0, y0)

# Kt qu
print('Solution (x, y) = (%f, %f), cost = %f, obtained after %d iterations' % (x[-1], y[-1], cost(x[-1], y[-1]), it))
````

Di y l mt v d v cch p dng Gradient Descent cho mt hm nhiu bin khc:

Hm mc tiu: $f(\mathbf{x}) = x_1^2 + 3x_2^2 - 2x_1x_2$

Gradient ca hm mc tiu: $\nabla f(\mathbf{x}) = \begin{bmatrix} 2x_1 - 2x_2 \\ 6x_2 - 2x_1 \end{bmatrix}$

Thut ton GD cho hm hai bin:

1. Chn gi tr ban u cho vector $\mathbf{x}$: $\mathbf{x}_0$.
2. Lp qua cc bc sau cho n khi tiu ch dng (v d: gradient nh hn mt ngng nht nh):
   - Tnh gradient ti im hin ti: $\nabla f(\mathbf{x}_i)$.
   - Cp nht vector $\mathbf{x}$ theo hng m ca gradient:
     $$\mathbf{x}_{i+1} = \mathbf{x}_i - \eta \nabla f(\mathbf{x}_i)$$
   - Tng s lng lp ln: $i = i + 1$.

Di y l v d m Python  ti u ha hm ny:



````python
import numpy as np
import matplotlib.pyplot as plt

def gradient(x):
    grad = np.array([2 * x[0] - 2 * x[1], 6 * x[1] - 2 * x[0]])
    return grad

def cost(x):
    return x[0]**2 + 3 * x[1]**2 - 2 * x[0] * x[1]

def gradient_descent(eta, x0, max_iter=100, epsilon=1e-5):
    x = [x0]
    for i in range(max_iter):
        grad = gradient(x[-1])
        x_new = x[-1] - eta * grad
        x.append(x_new)
        if np.linalg.norm(grad) < epsilon:
            break
    return x

# p dng GD  ti u ha hm f(x)
x0 = np.array([3, 4])  # Gi tr khi to (vector)
eta = 0.1  # Learning rate
x_opt = gradient_descent(eta, x0)

# Kt qu
final_x = x_opt[-1]
final_cost = cost(final_x)
iterations = len(x_opt) - 1
print('Solution (x, y) =', final_x)
print('Cost =', final_cost)
print('Obtained after', iterations, 'iterations')

# V biu 
x_range = np.linspace(-5, 5, 100)
y_range = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_range, y_range)
Z = X**2 + 3 * Y**2 - 2 * X * Y
contours = plt.contour(X, Y, Z, levels=20, colors='gray')
plt.clabel(contours, inline=1, fontsize=10)
x_path = [point[0] for point in x_opt]
y_path = [point[1] for point in x_opt]
plt.plot(x_path, y_path, marker='o', markersize=5, label='GD path', color='blue')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Gradient Descent Optimization for f(x) = x1^2 + 3x2^2 - 2x1x2')
plt.legend()
plt.grid(True)
plt.show()
````
